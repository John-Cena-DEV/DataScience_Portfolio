```python
import pandas as pd

# Example DataFrame
df = pd.read_csv("/content/Dataset.csv")



# Convert "TRUE"/"FALSE" strings to real booleans
for col in ['1 Month','2 Month','3 Month','4 Month','5 Month',
                  '6 Month','7 Month','8 Month','9 Month','10 Month',
                  '11 Month','12 Month']:
    df[col] = df[col].astype(str).str.upper().map({'TRUE': True, 'FALSE': False})

# Count number of TRUE per column
true_counts = df[['1 Month','2 Month','3 Month','4 Month','5 Month',
                  '6 Month','7 Month','8 Month','9 Month','10 Month',
                  '11 Month','12 Month']].sum()

print(true_counts)
```
